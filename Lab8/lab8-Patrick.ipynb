```python
#Question 1 and 2
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn import tree

# Use gini measure quality of split and build the decision tree. Record the model accuracy.
dp = pd.read_csv("heart-disease-dataset2.csv")

for col in dp.columns:
    dp = dp[dp[col] != '?']

x_col = [col for col in dp.columns if col != 'result']

X = dp[x_col]

Y = dp['result'].values

ds_gini = DecisionTreeClassifier(criterion="gini", random_state = 2)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

ds_train = ds_gini.fit(X_train, Y_train)

Y_pred = ds_train.predict(X_test)

dot_data = export_graphviz(ds_train, out_file=None, feature_names=x_col, class_names=['0','1','2','3','4'], rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

accuracy = accuracy_score(Y_test, Y_pred)

print(f'Accuracy Score: {round(accuracy * 100,2)} %')

Image(graph.create_png())

```

```python
#Question 5 and 6
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn import tree

# Use gini measure quality of split and build the decision tree. Record the model accuracy.
dp = pd.read_csv("heart-disease-dataset2.csv")

for col in dp.columns:
    dp = dp[dp[col] != '?']

x_col = [col for col in dp.columns if col != 'result']

X = dp[x_col]

Y = dp['result'].values

ds_gini = DecisionTreeClassifier(criterion="entropy", random_state = 2)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

ds_train = ds_gini.fit(X_train, Y_train)

Y_pred = ds_train.predict(X_test)

dot_data = export_graphviz(ds_train, out_file=None, feature_names=x_col, class_names=['0','1','2','3','4'], rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

accuracy = accuracy_score(Y_test, Y_pred)

print(f'Accuracy Score: {round(accuracy * 100,2)} %')

Image(graph.create_png())

```
